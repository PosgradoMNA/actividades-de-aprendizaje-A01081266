#Evaluación de modelos

*nos dicen que tan bien funciona nuestro modelo

*usamos datos para entrenarlo y para probarlo, generalmente se usa un 70%-30%

*se puede usar el Train test split Function de Sci kit

*la salida de la funcion son x_train, y_train, x_test, y_test

*los errores del modelos nos van a decir que tan bien entrenado esta el modelo, pudiento tener mejor accuracy, pero poca precision

*se puede usar k-fold para hacer cross validation

#overfitting, underfitting and model selection

* un modelo subentrenado va a tener muchos errores

* un modelo sobre entrenadotiene muy buen desempeño con los datos de entrenamiento, pero su desempeño con datos de prueba es mucho menor

#Ridge regression

* previene el sobreentrenamiento con la ayuda del parametro alpha

* alpha de valor alto va a reducir los coeficientes a cero y puede llegar a subentrenar el modelo

* si alpha es cero, el modelo se va a sobre entrenar

* para seleccionar un alpha se puede usar cross validation

#Grid Search

* nos permite escanear con muchos parametros con pocas lineas de codigo

* estos calores se llaman hyperparametros

* con ayuda de R2 podemos encontrar cual es el mejor valor para el modelo
